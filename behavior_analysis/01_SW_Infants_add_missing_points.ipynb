# Import packages

```python
import pandas as pd
import math
import numpy as np
import os
import glob
frames_per_sec = 30
```

```python
#directory where behavior videos are located
video_directory = '01_Raw_videos'
Adult_model_directory = '02_Adult_model'
Infant_model_directory = '03_Infants_model'
output_directory = '04_Infants_model_adjusted'
file_location = os.path.join(video_directory, '*.mp4')

# create list of filenames for every video to be processed
# this list is the input for the video processing function
filenames = glob.glob(file_location)
videoname = []
for f in filenames:
    
    # establish name for output file from the input filename
    x = os.path.split(f)
    x = x[1].split('.mp4')
    x = x[0]
    videoname.append(x)
videoname
```

# Import the output from 'refine trackelts' for Infants

```python
#Import the H5 file for Adults
#Import the H5 file for Adults
Adult_file = os.path.join(Adult_model_directory, x)
Adult_var = glob.glob(Adult_file + '*')
Adults = pd.read_hdf(Adult_var[0])
Adults = Adults.droplevel(level=0, axis=1)
Adults = Adults.apply(pd.to_numeric)
# The Adult DLC dataframe was created using single animal DLC so there is no multiindex level cooresponding to 'individuals'
# Must make the indexes match before we can merge.
Adults['individuals'] = 'Adult'
Adults = Adults.set_index('individuals', append=True).unstack('individuals')
Adults = Adults.reorder_levels([2,0,1], axis=1)
# Set dimesnions of dataframe so that all of TEST is visible when printed
#pd.set_option('display.max_rows', 500)
#pd.set_option('display.max_columns', 50)
```

```python
#Import the H5 file for Infants
Infant_file = os.path.join(Infant_model_directory, x)
Infant_var = glob.glob(Infant_file + '*')
Infants = pd.read_hdf(Infant_var[0])
Infants = Infants.droplevel(level=0, axis=1)
Infants = Infants.apply(pd.to_numeric)
Infant_bodyparts = Infants.columns.get_level_values(level=1)
Infant_bodyparts = set(Infant_bodyparts.tolist())
# Set dimesnions of dataframe so that all of TEST is visible when printed
#pd.set_option('display.max_rows', 500)
#pd.set_option('display.max_columns', 30)
```

```python
Animals = Adults.join(Infants, how='outer')
# Set dimesnions of dataframe so that all of TEST is visible when printed
#pd.set_option('display.max_rows', 12200)
pd.set_option('display.max_columns', 70)
Animals
```

# Fill in missing datapoints

```python
# Sam, place the code for filling in missing points here.
# Cut out anything that is for calculating latency or other behavioral measurements
# This notebook run start to finish should be able to:
# read in a .h5 file, extrapolate missing points, and then save an edited .h5 file

# Create various lists to add different sets of data to (their respective purposes will become clear in the following lines)
columns_likelihood = []
columns_name = []
columns_disappear = []
columns_reappear = []
columns_end_missing = []
columns_appear = []

# Iterate through the columns in TEST
for column in Infants.columns:
    
    # Add the likelihood column names to the columns_likelihood list, then skip them
    if 'likelihood' in column:
        columns_likelihood.append(column)
        continue
        
    # For the rest of the columns (those that actually show points), add the column names to the columns_name list
    else:
        columns_name.append(column)
        
    # Create lists for disappear (the last point at which a point is visible), reappear (the first point at which a point)
    # becomes visible after being missing, end_missing (the last point that is visible in the case that that point stays)
    # missing for the rest of the video), and start_missing (the first point that is visible in the case that that point
    # is missing at the start of the video)
    disappear = []
    reappear = []
    end_missing = []
    start_missing = []
    
    # Loop through the rows in TEST
    for i in range(len(Infants)):
        
        # Check if each point is null
        if pd.isnull(Infants.at[i, column]):
            
            if i - 1 not in Infants[column].index:
                continue
            
            # If null, check if the point before it is null, and if not, add that point to the disappear list
            if pd.notnull(Infants.at[i - 1, column]):
                disappear.append(i - 1)
                
            # To avoid indexing errors, check if the point ahead of each null point is in the index. If the end of the
            # index is reach, that means that the point that the last point that was added to the disappear list actually
            # belonged in the end_missing list, so add that point to the end_missing list and remove it from the disappear
            # list
            if i + 1 not in Infants[column].index:
                end_missing.append(disappear[-1])
                disappear.pop(-1)
                continue
            # If the point following the null point is not null, and if it is not, add that point to the reappear list
            elif pd.notnull(Infants.at[i + 1, column]):
                reappear.append(i + 1)
                
    # If the first value in the reappear list is smaller than the first value in the disappear list, that means that the
    # point was invisible at the start of the video and that the first value in the reappear list actually belongs in the
    # start_missing list
    if reappear == [] and disappear == []:
        continue
    if reappear[0] < disappear[0]:
        columns_appear.append(reappear[0])
        reappear.pop(0)
    
    # Add each of the lists to the corresponding columns_(insert corresponding list name) list so that each index in the
    # new list is a list of the necessary frames for each column
    columns_disappear.append(disappear)
    columns_reappear.append(reappear)
    columns_end_missing.append(end_missing)

# Iterate through the columns_name list
for i in range(len(columns_name)):
    
    # Iterate through each list in the columns_disappear list
    for j in range(len(columns_disappear[i])):
        # Check if each list in columns_disappear is empty, and if so skip
        if columns_disappear[i] == []:
            continue
            
        # Find the difference between the value each point in each list of columns_reappear with the corresponding point in
        # the corresponding list of columns_disappear
        if (i + 2) <= len(columns_name):
            difference_x = Animals.at[columns_reappear[i][j], columns_name[i]] - Animals.at[columns_disappear[i][j], columns_name[i]]
            difference_y = Animals.at[columns_reappear[i][j], columns_name[i + 1]] - Animals.at[columns_disappear[i][j], columns_name[i + 1]]
        
            # Code for when the adult picks up the infant
            if (abs(difference_x) > 100 and 'x' in columns_name[i]) or (abs(difference_y) > 100 and 'y' in columns_name[i + 1]):

                v1 =  Animals.at[columns_disappear[i][j], columns_name[i]]  - Adults.at[columns_disappear[i][j], 
                ('Adult', 'head_center', 'x')]
                v2 = Animals.at[columns_disappear[i][j], columns_name[i + 1]]  - Adults.at[columns_disappear[i][j], 
                ('Adult', 'head_center', 'y')]

                x1 = Animals.at[columns_disappear[i][j], ('Adult', 'neck', 'x')]
                y1 = Animals.at[columns_disappear[i][j], ('Adult', 'neck', 'y')]
                x2 = Animals.at[columns_disappear[i][j], ('Adult', 'head_center', 'x')]
                y2 = Animals.at[columns_disappear[i][j], ('Adult', 'head_center', 'y')]

                rise1 = (y2 - y1)
                run1 = (x2 - x1)
                m1 = rise1/run1
                for k in range(columns_disappear[i][j] + 1, columns_reappear[i][j]):

                    end_diff_head_x = abs(Animals.at[columns_reappear[i][j], columns_name[i]] - Animals.at[int(k), ('Adult', 'head_center', 'x')])
                    end_diff_head_y = abs(Animals.at[columns_reappear[i][j], columns_name[i + 1]] - Animals.at[int(k), ('Adult', 'head_center', 'y')])


                    # For when the adult gets near the point where they drop the infant
                    if end_diff_head_x < 30 and end_diff_head_y < 15:

                        for l in range(k, columns_reappear[i][j]):

                            Animals.iat[int(l), int(Animals.columns.get_loc(columns_name[i]))] = Animals.at[columns_reappear[i][j],
                            columns_name[i]]
                            Animals.iat[int(l), int(Animals.columns.get_loc(columns_name[i + 1]))] = Animals.at[columns_reappear[i][j],
                            columns_name[i + 1]]
                        break

                    x3 = Animals.at[k, ('Adult', 'head_center', 'x')]
                    y3 = Animals.at[k, ('Adult', 'head_center', 'y')]
                    x4 = Animals.at[k, ('Adult', 'nose', 'x')]
                    y4 = Animals.at[k, ('Adult', 'nose', 'y')]

                    rise2 = (y4 - y3)
                    run2 = (x4 - x3)
                    m2 = rise2/run2

                    angle1 = np.arctan(m1)
                    angle2 = np.arctan(m2)
                    angle_change = angle2 - angle1

                    if run1 < 0:
                        angle_change -= np.pi
                    if run2 < 0:
                        angle_change += np.pi

                    v1_new = v1 * np.cos(angle_change) - v2 * np.sin(angle_change)
                    v2_new = v1 * np.sin(angle_change) + v2 * np.cos(angle_change)

                    Animals.iat[int(k), int(Animals.columns.get_loc(columns_name[i]))] = v1_new + Animals.at[int(k), ('Adult', 'head_center', 'x')]
                    Animals.iat[int(k), int(Animals.columns.get_loc(columns_name[i + 1]))] = v1_new + Animals.at[int(k), ('Adult', 'head_center', 'y')]
                
                continue
        
        if pd.notnull(Animals.at[columns_disappear[i][j] + 1, columns_name[i]]):
            continue
            
        # Find the difference in frames between each point in each list of the columns_reappear with the corresponding point
        # in the corresponding list of columns_disappear
        frames_difference = columns_reappear[i][j] - columns_disappear[i][j]
        
        # Find the average between the disappear and reappear points by dividing difference by frames difference
        gap_average = difference_x / frames_difference
        
        # Start a counter at one that represents the number of frames after each point in each list of columns_disappear
        position = 1
        
        # Iterate through each point between each frame number in each list in columns_disappear and columns_reappear
        for k in range(columns_disappear[i][j] + 1, columns_reappear[i][j]):
            
            # Calculate the value that each missing point should be based on the gap_average value times the position value
            # for that particular frame, then set that point to that value and increment position by 1
            value = Animals.at[columns_disappear[i][j], columns_name[i]] + gap_average * position
            Animals.iat[int(k), int(Animals.columns.get_loc(columns_name[i]))] = value
            position += 1
        
            
# Loop through columns_name again (this time to add in points that were missing at the end or start of the video)
for i in range(len(columns_name)):
    
    # Iterate through each list in columns_end_missing
    for j in range(len(columns_end_missing[i])):
        
        # Check if each list in columns_end_missing is empty, and skip if it is
        if columns_end_missing[i] == []:
            continue
        # Set value to the value of the point in columns_end_missing for that column, and just like before set a position
        # variable at 1
        value = Animals.at[columns_end_missing[i][j], columns_name[i]]
        position = 1
        
        # Start a forever loop
        while True:
            
            # Check if each particular frame following the value in each list in columns_end_missing is not in the index
            # and break if that is true
            if columns_end_missing[i][j] + position not in Animals[columns_name[i]].index:
                break
            
            # Set each point following the frame in each list of columns_end_missing to value, then increment position by 1
            Animals.iat[columns_end_missing[i][j] + position, int(Animals.columns.get_loc(columns_name[i]))] = value
            position += 1
            
# For columns_disappear, columns_reappear, columns_start_missing and columns_end_missing, create a likelihood columns for
# each that only includes every other list in each of these list
# Basically, each pair in these starting lists includes an x and y point, and both of these points are missing at the same
# time along with a corresponding likelihood value, so we only need each of these frames once in the likelihood lists for
# the single corresponding likelihood value
columns_likelihood_disappear = []
columns_likelihood_reappear = []
columns_likelihood_end_missing = []
for i in range(0, len(columns_name), 2):
    columns_likelihood_disappear.append(columns_disappear[i])
    columns_likelihood_reappear.append(columns_reappear[i])
    columns_likelihood_end_missing.append(columns_end_missing[i])
    
# Loop through the likelihood columns
for i in range(len(columns_likelihood)):
    
    # Loop through each list of columns_likelihood_disappear
    for j in range(len(columns_likelihood_disappear[i])):
        
        # Check if each list inside columns_likelihood_disappear is empty (this also means that the corresponding list in
        # columns_likelihood_reappear is empty), and skip if so
        if columns_likelihood_disappear[i] == []:
            continue
        
        # Loop through all the points between each frame designated in each list in columns_likelihood_disappear and
        # columns_likelihood reappear, setting the likelihood value to 0.01
        for k in range(columns_likelihood_disappear[i][j] + 1, columns_likelihood_reappear[i][j]):
            Animals.iat[int(k), int(Animals.columns.get_loc(columns_likelihood[i]))] = float("0.01")
            
    # Loop through each list of columns_likelihood_end_missing
    for j in range(len(columns_likelihood_end_missing[i])):
        # Check if each list inside columns_likelihood_end_missing is empty, and skip if so
        if columns_likelihood_end_missing[i] == []:
            continue
            
        # Set a position variable to 1, then start a forever loop
        position = 1
        while True:
            
            # If the next value in the iteration isn't in the index, break
            if columns_likelihood_end_missing[i][j] + position not in Animals[columns_name[i]].index:
                break
                
            # Set each point in the likelihood column to 0.01, and increment the position variable by 1
            Animals.iat[columns_likelihood_end_missing[i][j] + position, int(Animals.columns.get_loc(columns_likelihood[i]))] = float("0.01")
            position += 1
            
# remove Adult bodyparts and return only thefully modified Infant points            
Animals = Animals.drop(['Adult'], axis=1)

Animals
```
